## distplot

O distplot utiliza a estrutura de um histograma e pode-se observar 1 variável: ou seja, uma coluna da distribuição de dados

Para visualizar o distplot, utiliza-se a seguinte sintaxe:



```
sns.distplot(a=seu_dataframe['coluna_desejada'], kde=False)
```



````python
#Observar os valores do comprimento das sépalas
sns.distplot(a=iris_data['SepalLengthCm'], kde=False)
````

Mas o que são sépalas?

picture)

Agora que sabemos as partes da flor, observemos a largura da sépalas

````python
sns.distplot(a=iris_data['SepalWidthCm'], kde=False)
````

Dessa forma, podemos utilizar esse mesmo método para as demais informações que quisermos

## kdeplot


O kdeplot (Kernel Density Estimate (KDE) Plot) é semelhante ao distplot, mas ao invés de ser "quadradão", ele é mais arredondado, possui curvas mais suvaes

Para visualizar os dados utilizando kdeplot, utiliza-se a seguinte linguagem:

```
sns.kdeplot(data=seu_dataframe['coluna_desejada'], shade=True)
```



````python
sns.kdeplot(data=iris_data['PetalLengthCm'], shade=True)
````

## jointplot

O jointplot é como o kdeplot mas com 2 variáveis

Pode-se ser usado na forma de hisograma ou de KDE

Sintaxe para usar o jointplot:


Como histograma:
```
sns.jointplot(x=seu_dataframe['coluna_desejada'], y=seu_dataframe['outra_coluna'])
```

Como KDE:
```
sns.jointplot(x=seu_dataframe['coluna_desejada'], y=seu_dataframe['outra_coluna'], kind='kde')
```


````python
#Forma de histograma
sns.jointplot(x=iris_data['PetalLengthCm'], y=iris_data['SepalWidthCm'])
````

````python
#Forma de KDE
sns.jointplot(x=iris_data['PetalLengthCm'], y=iris_data['SepalWidthCm'], kind='kde')
````

## linecharts

Para demonstrar a utilização de gráficos de linhas, utilizaremos os dados de quantidade de acesso a 5 músicas específicas no Spotify, em uma janela de tempo específica

````python
spotify_url = 'https://raw.githubusercontent.com/gabrielMelo42/Workshop-Data-Visualization/main/spotify.csv'
spotify_data = pd.read_csv(spotify_url)
spotify_data.head()
````

Como Shape of You foi lançada antes das outras músicas, utilizando head() não veremos dados sobre as demais; por isso, utilizaremos tail(), que mostra as últimas 5 linhas ao invés das 5 primeiras

````python
spotify_data.tail()
````

Para visualizar o linechart, utiliza-se a seguinte sintaxe:



```
sns.lineplot(data=seu_dataframe)
```



````python
sns.lineplot(data=spotify_data)
````

Vamos colocar um título no gráfico

````python
plt.pyplot.title("Músicas que tá todo mundo ouvindo em 2017-2018")
sns.lineplot(data=spotify_data)
````

O gráfico está um pouco pequeno e não tão bom de visualizar; podemos mudar seu tamanho

````python
#Escolhe-se o tamanho do gráfico
plt.pyplot.figure(figsize=(10,5))

plt.pyplot.title("Músicas que tá todo mundo ouvindo em 2017-2018")
sns.lineplot(data=spotify_data)
````

Podemos inclusive plotar apenas 1 das músicas

# Visualização de dados "Relationship"


Uma das coisas que mais queremos encontrar utilizando a Ciência de Dados, é a relação entre 2 ou mais features diferentes de um conjunto de dados, porque assim poderemos pensar em possíveis soluções para um problema ou tomar decisões que tragam um resultado positivo para a empresa, por exemplo

## barchart

No gráfico de barra (com formato de histograma), o eixo x tem o valor dos dados enquanto o eixo y possui a categoria ddos dados

Vamos utilizar um conjunto de dados sobre média nos atrasos de voos em cada mês por empresas áreas diferntes nos Etados Unidos

As linhas correspondem aos meses, as colunas à empresa área e os dados a média de atraso de voo em minutos que uma determinada empresa atrasou em um dos meses do ano (há 12 linhas porque 1 ano tem 12 meses)

````python
flight_url ='https://raw.githubusercontent.com/gabrielMelo42/Workshop-Data-Visualization/main/flight_delays.csv'
flight_data = pd.read_csv(flight_url)
flight_data.head()
````

Sintaxe:



```
sns.barplot(x=seu_dataframe, y=seu_dataframe['uma_coluna_especifica'])
```



Mostraremos a média de atrasos em cada mês para uma das empresas:

Mais uma vez o gráfico encontra-se um pouco p

````python
plt.pyplot.figure(figsize=(10,5))
plt.pyplot.title('Média de atrasos de voos da empresa área NK, por mẽs')
sns.barplot(x=flight_data.index, y=flight_data['NK'])
#O eixo y corresponde ao número de minutos que a emprsa atrasou, enquanto o eixo x corresponde ao mês do ano
plt.pyplot.ylabel('Atraso em minutos')
````

Podemos fazer isso para cada empresa:

````python
plt.pyplot.figure(figsize=(12,4))
plt.pyplot.title('Média de atrasos de voos da empresa área AA, por mẽs')
sns.barplot(x=flight_data.index, y=flight_data['AA'])
plt.pyplot.ylabel('Atraso em minutos')
````

Vê-se que a empresa AA teve, em alguns meses, médias negativas, significando que a média foi de estarem adiantados em relação à hora prevista

## heatmap


Heatmap ou Mapa de calor são utilizados para calcular a "intensidade" de um determinado evento, de um determinado acontecimento, em que as regiões mais escuras tem maior "intensidade"


Utilizaremos o mesmo dataset que utilizamos no barchat

Sintaxe:



```
sns.heatmap(data=seu_dataframe, annot=True)

```



````python
plt.pyplot.figure(figsize=(14,7))
plt.pyplot.title('Média de atrasos para cada companhia aérea, por mês')

#O heatmap em si
sns.heatmap(data=flight_data, annot=True)

# O eixo x são as companhias aéreas enquanto o eixo y são os meses
plt.pyplot.xlabel('Companhia aérea')
````

Se não utilizamos o ```annot=True```, os valores não serão mostrados, apenas as cores que representam os valores, por isso é sempre importante colocar essa especificação

````python
plt.pyplot.figure(figsize=(14,7))
plt.pyplot.title('Média de atrasos para cada companhia aérea, por mês')

#O heatmap em si
sns.heatmap(data=flight_data)

# O eixo x são as companhias aéreas enquanto o eixo y são os meses
plt.pyplot.xlabel('Companhia aérea')
````

## scatterplot

O scatterplot ou gŕafico de dispersão representam 2 ou mais variáveis e os "pontos" de encontro dessas, utilizando um plano cartesiano e suas coordenadas



Vamos utilizar um novo dataset, sobre saúde e despesas de seguro:

````python
saude_url = 'https://raw.githubusercontent.com/gabrielMelo42/Workshop-Data-Visualization/main/insurance.csv'
saude_data = pd.read_csv(saude_url)
saude_data.head()
````

Vamos substituir os nomes das features para português

````python
saude_data.columns =['Idade', 'Sexo', 'IMC', 'Quantos filhos', 'Fuma?', 'Região', 'Despesas']
````

````python
saude_data.head()
````

Como utilizar scatterplot?



```
sns.scatterplot(x=seu_dataframe['uma determinada coluna'], y=seu_dataframe['uma_outra_determinada_coluna'])
```



````python
sns.scatterplot(x=saude_data['IMC'], y=saude_data['Despesas'])
````

Pode-se observar que vários indivíduos desse dataset de maior Indíce de Massa Corporal têm mais despesas com plano de saúde, fato que pode ser explicado por possuirem um maior risco de doenças crônicas

Assim, IMC e despesas com o plano de saúde são positivamente correlacionados


Pode-se também classificar os valores encontrados no gráfico a partir de outra feature, podendo entender melhor o conjunto de dados que possuimos:

Sintaxe:



```
sns.scatterplot(seu_dataframe['uma_determinada_coluna'], seu_dataframe['uma outra coluna], hue = seu_dataframe['label'])```



````python
sns.scatterplot(x=saude_data['IMC'], y=saude_data['Despesas'], hue=saude_data['Fuma?'])
````

Observa-se que não só o IMC tem influência sobre a quantidade de gastos em plano de saúde, mas também -e principalmente- o fato de fumar ou não: quase todos que fumam pagam mais que quase todos que não fumam

### regplot

Com o método regplot(), pode-se utilizar um scatteplot com linha de regressão:

````python
sns.scatterplot(x=saude_data['IMC'], y=saude_data['Despesas'])
````

````python
sns.regplot(x=saude_data['IMC'], y=saude_data['Despesas'])
````

O gráfico permite observar que os não fumantes tendem a pagar um pouco mais com o aumento do IMC, mas os fumantes pagam muito mais com o aumento do IMC

Utilizando duas linhas de regressão, uma para fumantes e uma para não fumantes, poder-se-á notar melhorar essas particularidades daqueles que fumam ou não:



### lmplot

````python
sns.scatterplot(x=saude_data['IMC'], y=saude_data['Despesas'], hue=saude_data['Fuma?'])
````

````python
sns.lmplot(x='IMC', y='Despesas', hue='Fuma?', data=saude_data)
````

Sintaxe:



```
sns.lmplot(x='uma_coluna', y='outra_coluna', hue='aquilo_que_voce_quer_observar', data=seu_dataframe)
```



Nota-se que a inclinação da linha de regressão para fumantes é mais acentuada que a linha de regressão para não fumantes, mostrando o que falamos anteriormente: os não fumantes tendem a pagar um pouco mais com o aumento do IMC, mas os fumantes pagam muito mais com o aumento do IMC

##swarmplot: o scatterplot para dados categóricos!

Os scatter plots observam as (co)relações entre 2 variáveis contínuas, mas se utiizaremos dados categóricos, utilizaremos o swarmplot

Uma das variávies categórias no nosso dataframe é se o indivíduo é fumante ou não (só existem essas 2 opções, sendo essas categóricas)

Comparemos essa feature com as quantidade de dinheiro que ele pagará com plano de saúde

Sintaxe:



```
sns.swarmplot(x=seu_dataframe['uma_feature_categorica'], y= seu_dataframe['a_coluna_de_interesse_do_dataset'])

```



````python
sns.swarmplot(x=saude_data['Fuma?'], y=saude_data['Despesas'])
````

No geral, os fumantes gastam mais dinheiro com plano de saúde que os não fumantes

Além disso, os clientes que pagam os maiores valores são os fumantes, assim como os indivíduos que pagam os menores valores são não fumantes. 

# Algumas formas interessantes de vizualizar dados

Essas formas interessantes de visualizar dados será baseada no artigo https://towardsdatascience.com/comparing-different-ways-of-displaying-categorical-data-in-python-ed8fabfb6661 do Towards Data Science

Usaremos também o código presente no artigo

Setamos o dataset, dataset esse que indica as áreas de cada continente do planeta Terra

````python
dct = {'Oceania': 8.5, 'Europa': 10, 'Antártida': 14.2, 'América do Sul': 17.8, 
       'América Central': 0.7, 'América do Norte': 24.2, 'África': 30.4, 'Asia': 44.6}
continents = list(dct.keys())
populations = list(dct.values())
````

Já conhecemos o barplot:

````python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,4))
plt.barh(continents, populations, color='slateblue')
plt.title('Área dos continentes', fontsize=27)
plt.xlabel('Área em milhões de km2', fontsize=19)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.tick_params(left=False)
sns.despine(left=True)
plt.show()
````

##Stem plot

````python
plt.figure(figsize=(10,4))
plt.hlines(y=continents, xmin=0, xmax=populations, color='slateblue')
plt.plot(populations, continents, 'o', color='slateblue')
plt.title('Área dos continentes', fontsize=27)
plt.xlabel('Área em milhões de km2', fontsize=19)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlim(0, None)
plt.tick_params(left=False)
sns.despine(left=True)
plt.show()
````

##Pie Chart ou Gráfico de Pizza

````python
plt.figure(figsize=(7,7))
plt.pie(populations, labels=continents, startangle=90, autopct='%1.0f%%', 
        textprops={'fontsize': 15})
plt.text(x=-1.15, y=1.5, s='Área dos continentes em porcentagem', fontsize=27)
plt.show()
````

Enquanto que nos dois primeiros gráficos mostrava-se a área quantitativa, em km quadrados, de cada continente, no Pie Chart mostra-se a área percentual de cada continente em relação à superfície terrestre -ou seja, da Terra sem contar O Oceano

Geralmente, recomenda-se não utilizar Pie Chart em DataViz, trouxe essa forma de visualizar os dados a título de informação

## Waffle chart

Um gráfico que lembra um waffle, funciona melhor quando temos um pequeno conjunto de dados discretos e estamos interessados na estrutura interna dos seus componentes

Há a necessidade de instalar e importar uma biblioteca do python chamada pywaffl

````python
!pip install pywaffle
````

````python
from pywaffle import Waffle
````

````python
fig = plt.figure(FigureClass=Waffle, figsize=(12,7), values=dct, columns=15, block_arranging_style='snake', legend=
                        {'bbox_to_anchor': (1.3, 1), 'fontsize':14, 'frameon': False})
plt.title('Àrea dos continentes', fontsize=33)
plt.show()
````
